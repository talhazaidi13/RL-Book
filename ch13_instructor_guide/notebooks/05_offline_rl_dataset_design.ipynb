This notebook creates a schema and support-audit template. It does not claim to implement a full pushing simulator.

```python
import pandas as pd
import numpy as np

required_fields = [
    "episode_id", "step", "observation", "action", "executed_action",
    "reward", "cost", "next_observation", "terminated", "truncated",
    "task_id", "object_id", "scene_id", "timestamp", "control_hz",
    "success", "failure_type", "safety_filter_activated",
    "contact_force", "behavior_policy_id"
]
pd.DataFrame({"required_field": required_fields})
```

```python
behavior_categories = pd.DataFrame([
    {"category": "success", "example": "Direct push to goal", "why_needed": "Defines successful support"},
    {"category": "failure", "example": "Missed contact", "why_needed": "Supports value discrimination"},
    {"category": "failure", "example": "Overshoot", "why_needed": "Teaches cost of excessive action"},
    {"category": "recovery", "example": "Re-approach after miss", "why_needed": "Supports closed-loop correction"},
    {"category": "recovery", "example": "Correct lateral drift", "why_needed": "Covers off-nominal states"},
])
behavior_categories
```

```python
# Synthetic action-support audit example.
rng = np.random.default_rng(5)
logged_actions = rng.normal(0, 0.25, size=(5000, 2))
candidate_actions = np.array([
    [0.1, 0.1],
    [0.4, -0.2],
    [1.2, 1.2],
])

def nearest_distance(candidate, dataset):
    return np.linalg.norm(dataset - candidate, axis=1).min()

for action in candidate_actions:
    print(action, "nearest logged-action distance =", nearest_distance(action, logged_actions))
```

## Instructor interpretation

Nearest-neighbor distance is only a simple diagnostic. It does not prove support in high-dimensional state-action space. A stronger audit conditions on similar observations and uses density, uncertainty, or representation-space diagnostics.
